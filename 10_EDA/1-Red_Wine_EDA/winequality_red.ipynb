## Wine Quality - Red Wine Dataset

The two datasets are related to red and white variants of the Portuguese "Vinho Verde" wine. Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

These datasets can be viewed as classification or regression tasks.  The classes are ordered and not balanced (e.g. there are many more normal wines than excellent or poor ones). Outlier detection algorithms could be used to detect the few excellent or poor wines. Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.

**Additional Variable Information**

Input variables (based on physicochemical tests):
-   1 - fixed acidity
-   2 - volatile acidity
-   3 - citric acid
-   4 - residual sugar
-   5 - chlorides
-   6 - free sulfur dioxide
-   7 - total sulfur dioxide
-   8 - density
-   9 - pH
-   10 - sulphates
-   11 - alcohol

Output variable (based on sensory data):
 
-   12 - quality (score between 0 and 10)

---

```python
# Countplot for wine quality
sns.countplot(x="quality",hue='quality', data=df, palette="viridis")
plt.title("Distribution of Wine Quality Ratings")
plt.show()
```

This confirms class imbalance: most wines are rated 5 or 6, while very few are rated at the extremes (3, 4, 8). This is important context for later modeling.

---

```python
# to see distribution for each and every feature in dataset. ofcourse quality is discrete feature and we should use barplot for it. But its EDA

# Melt dataset to long format
df_melted = df.melt(var_name="Feature", value_name="Value")

# Plot histograms with FacetGrid
g = sns.FacetGrid(df_melted, col="Feature", col_wrap=4, sharex=False, sharey=False)
g.map(sns.histplot, "Value",bins=20, color="steelblue")
plt.show()
```

This gives a quick overview of how each feature is distributed. You can spot skewed features (like residual sugar or sulphates) and approximately normal ones (like pH).

---

```python
df_PDF = df.melt(var_name='Feature', value_name='Value')

# visualize distribution using ked (Kernal Density Estimator)
plt.Figure(figsize=(10, 6))
g = sns.FacetGrid(df_PDF, col='Feature', col_wrap=4, sharex=False, sharey=False)
g.map(sns.kdeplot, "Value", color='darkblue')
```

Instead of histograms, we use KDE (Kernel Density Estimation) which shows smooth probability density curves. This makes it easier to compare shapes of distributions.
For example: alcohol is right-skewed, pH is bell-shaped, and quality has spikes (discrete).
