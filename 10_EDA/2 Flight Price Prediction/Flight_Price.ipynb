## Flight Price Prediction Dataset Feature Engineering

### FEATURES

The various features of the cleaned dataset are explained below:

- 1) Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines.
- 2) Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
- 3) Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities.
- 4) Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels.
- 5) Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
- 6) Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
- 7) Destination City: City where the flight will land. It is a categorical feature having 6 unique cities.
- 8) Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy.
- 9) Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.
- 10)Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.
- 11) Price: Target variable stores information of the ticket price.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
# Lets handle the date column first
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
# Drop the Date column since we don't need it anymore 

df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.head()
```

```python
## Let's handle Arrival_Time
## Remove the Date and Month from the column as we already have it. 
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
## Extract Hour and Minute and Create new Columns
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
# convert the time to int 
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
```

```python

## Drop the Arrival_Time column we dont need it anymore.
df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head(4)
```

```python
### Lets handle Dep_Time

## Create 2 new Columns from the Dep_Time : 'Dep_hour' and 'Dep_min'
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
## Convert new Columns 'Dep_hour' and 'Dep_min' to int type
df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_min'] = df['Dep_min'].astype(int)
```

```python
df.info()
```

```python
## Drop the Dep_Time Column as we dont need it anymore
df.drop('Dep_Time', axis=1, inplace=True)
```

```python
df.head()
```

```python
## Lets handle Total Stops columns
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
# Since its a categorical column we can use mode imputation to fill that single nan value with mode
df['Total_Stops'].mode()
```

```python
df['Total_Stops'].unique()
```

```python
## We'll Use ordinal Encoding to replace all this categories with numbers
df['Total_Stops']= df['Total_Stops'].map({
    'non-stop':0,
    '1 stop':1,
    '2 stops':2,
    '3 stops':3,
    '4 stops':4,
    np.nan:1     # replacing nan value with mode
})
```

```python
df.head()
```

```python
## Lets drop route as we already have 2 seperate columns for Source and Destination
df.drop('Route', axis=1, inplace=True)
```

```python
df.head()
```

```python
## Lets handle Duration column and convert
import re
def convert_to_minutes(duration_str):
    # Create a pattern that can extract hours and minutes
    pattern = r"(?:(\d+)h)?(?:\s*(\d+)m)?"
    # match the pattern against the Duration fields 
    match = re.match(pattern, duration_str)
    # handle CornerCase
    if not match:
        return np.nan
    # separate the hour and minute and convert them into int and 0 if missing
    hours = int(match.group(1)) if match.group(1) else 0
    minutes = int(match.group(2)) if match.group(2) else 0
    # Calculate the minutes and return them
    return hours * 60 + minutes

df['Duration_min'] = df['Duration'].apply(convert_to_minutes)
```

```python
## Now drop the Duration column we don't need it
df.drop('Duration', axis=1, inplace=True)
```

```python
## Lets handle 'Airline' 'Source' 'Destination' Columns using OneHotEncoding
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
df_encoded = pd.DataFrame(encoder.fit_transform(df[['Airline', 'Source', 'Destination']]).toarray(), columns=encoder.get_feature_names_out())
```

```python
## Merge the df_encoded and df
df = pd.concat([df, df_encoded], axis=1)
```

```python
df.head()
```

```python
# Drop The Arrival, Source, and Destination Columns as we don't need them
df.drop(['Airline', 'Destination', 'Source'], axis=1, inplace=True)

```
