```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#-------------------------------------
# 1. Visulaize Line (2D)
#-------------------------------------
x = np.linspace(0, 10, 100)   # input feature
w = 2     # slope
b = 3     # intercept
y = w * x + b    # equation of line

plt.figure(figsize=(6, 4))
plt.plot(x, y, label='y = wx + b', color='blue')
plt.xlabel('x (feature)')
plt.xlabel('y (prediction)')
plt.title('Equation of a Line (2D)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#--------------------------------------
# 2. Visualize Plane (3D)
#--------------------------------------
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Generate grid of points
x1 = np.linspace(0, 10, 30)
x2 = np.linspace(0, 10, 30)
x1, x2 = np.meshgrid(x1, x2)

# Equation of plane: y = w1*x1 + w2*x2 +b
w1, w2, b = 1, 2, 5
y = w1 * x1 + w2 * x2 + b

# Plot surface
ax.plot_surface(x1, x2, y, cmap='viridis', alpha=0.8)

ax.set_xlabel('x1 (feature1)')
ax.set_ylabel('x2 (feature2)')
ax.set_zlabel('y (prediction)')
ax.set_title('Equation of a Plane (3D)')
plt.show()
```
