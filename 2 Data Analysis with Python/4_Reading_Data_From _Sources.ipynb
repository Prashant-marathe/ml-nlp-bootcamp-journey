## Reading Data From Sources

```python
import pandas as pd
from io import StringIO

data =  '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'

# Read From Json 
df = pd.read_json(StringIO(data))
df

# convert dataframe back to json
df.to_json(orient='records')   # indexing record based orient is the index
df.to_json(orient='index')     # indexing record on index
```

```python
# Reading HTML documents from web 
url = 'https://www.fdic.gov/bank-failures/failed-bank-list'
df = pd.read_html(url)
df[0]
df[0].to_excel('bank_data.xlsx')
```

```python
# read excel file
df = pd.read_excel('bank_data.xlsx')
df.head()
 
```

```python
''' Pickle file : A pickle file, usually with a .pkl or .pickle extension, is a file that stores a serialized Python object. Pickling is the process of converting a Python object hierarchy into a byte stream, and unpickling is the inverse process of converting that byte stream back into an object. This allows you to save a complex object, like a machine learning model or a custom class instance, and reconstruct it later with its exact state and data. ''' 

df_excel = pd.read_excel('bank_data.xlsx')

df_excel.to_pickle('df_excel')

pd.read_pickle('df_excel')
```
