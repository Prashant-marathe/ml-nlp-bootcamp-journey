## Missing Values
Missing values occur in dataset when some of the information is not stored for a variable. There are 3 Mechanisms:

**Missing Completely at Random (MCAR):**

Missing Completely at Random (MCAR) is a type of mechanism in which the probability of the value being missing is unrelated to both the observed and the missing data. In other words, if the data is MCAR, the missing values are randomly distributed throughtout the dataset, and there is no systematic reason for why they are missing.

For Example, In a survey about the prevalence of a certain disease, the missing data might be MCAR if the survey participants with missing values for certain questions were selected randomly and their missing responses are not related to their disease status or any other variables measured in the survey.

**Missing at Random (MAR):**

Missing at Random (MAR) is a type of missing data mechanism in which the probability of a value being missing depends only on the observed data, but not the missing data itself. In other words, if the data is MAR, the missing values are systematically related to the observed data, but not to the missing data. Here are a few example of missing at random:

Income data: Suppose you are collecting income data from a group of people, but osme perticipants choose not to report their income. If the decision to report or not report is related to the participant's age or gender, but not to their income level, then the data is missing at random.

Medical data: Suppose you are collecting medical data on patients, including their blood pressure, but some patients do not report their blood pressure. if the patients who do not report their blood pressure are more likely to be younger or have a healthier lifestyles, but the missingness is not related to their actual blood pressure values,then the data is missing at random.

**Missing Data Not at Random (MNAR):**

It is a type of missing data mechanism where the probability of missing values depends on the value of missing data itself. In other words, if the data is MNAR, the missingness id not random and is dependent on unobserved or unmeasured factors that are associated with the missing values.

For Example, suppose you are collection data on the income and job satisfaction of employess in a company. if employees who are less satisfied with their jobs are more likely to refuse to report their income, then the data is not missing at random. In this case, the missingness is dependent on job satisfaction, which is not directly observed or measured.

## Imputing Missing Values using Techniques:

### 1. Mean Value Imputation : Filling missing values in a column with the average of that column. Works well when we have normally distributed data.

```python
# displot 
sns.displot(df['age'])

# histogram with KDE (kernal Density Estimator)
sns.histplot(df['age'], kde=True)

#* Mean Value Imputation
#* Lets fill missing values in the age column with mean of the column (Mean Value Imputation)
df['Age_Mean'] = df['age'].fillna(df['age'].mean())
df[['Age_Mean', 'age']]
```

### 2. Median Value Imputation : Filling missing values in a column with median of that column. Useful when we have left-skewed or right-skewed dtstribution.

```python
df['Age_median'] = df['age'].fillna(df['age'].median())
df[['Age_median','Age_Mean', 'age']]
```

### 3. Mode Value Imputation: Filling missing values in a categorical column with mode. Specifically for categorical columns

```python
# We have a categorical column name embarked which likely follows MCAR mechanism
df[df['embarked'].isnull()]
```

```python
df['embarked'].unique()
# categorical column with ['S', 'Q', 'C', nan] categories
```

```python
mode_value = df[df['embarked'].notna()]['embarked'].mode()[0]
```

```python
df['embarked_mode'] = df['embarked'].fillna(mode_value)
```

```python
df[['embarked_mode', 'embarked']]
```

```python
df['embarked_mode'].isnull().sum()
```

```python
df['embarked'].isnull().sum()
```

### 3.Random value imputation:
method for filling in missing data by replacing NaN values with randomly sampled existing (non-missing) values from the same column, preserving the original distribution

```python
import numpy as np
nan_mask = df['age'].isna()
df.loc[nan_mask, 'age'] = np.random.choice(df['age'].dropna(), size=nan_mask.sum())
df['age'].isnull().sum()
```
