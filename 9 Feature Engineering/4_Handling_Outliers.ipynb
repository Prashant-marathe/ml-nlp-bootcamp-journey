### 5 Number Summary and Box Plot (Handling Outliers)
1.Minimum

2.Q1

3.Median

4.Q3

5.Maximum

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

number_lst = [7, 8, 5, 6, 7, 9, 10, 6, 7, 8, 50, 55]
minimum, Q1, median, Q3, maximum = np.quantile(number_lst, [0, 0.25, 0.5, 0.75, 1.0])

print(f'{minimum}, {Q1}, {median}, {Q3}, {maximum} ')

# calculate IQR
IQR = Q3 - Q1

print(f'IQR={IQR}')

# calculate lower fense and higher fense
lower_fense = Q1 - 1.5 * (IQR)
higher_fense = Q3 + 1.5 * (IQR)

print(f'lower_fense: {lower_fense} and higher_fense: {higher_fense}')

# lets use boxplot to visualize the outliers
sns.boxplot(number_lst)
plt.title('Boxplot: numbers_lst (outliers at 50, 55)')
```

As we can see most of the values in numbers_lst range between 5-10 and the list has 2 outliers 50, and 55
